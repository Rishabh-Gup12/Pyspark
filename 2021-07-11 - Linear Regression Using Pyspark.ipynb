## Overview

This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. [DBFS](https://docs.databricks.com/user-guide/dbfs-databricks-file-system.html) is a Databricks File System that allows you to store data for querying inside of Databricks. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

This notebook is written in **Python** so the default cell type is Python. However, you can use different languages by using the `%LANGUAGE` syntax. Python, Scala, SQL, and R are all supported.

```
# File location and type
file_location = "/FileStore/tables/tips-2.csv"
file_type = "csv"

df=spark.read.csv(file_location,header=True, inferSchema= True)
```

```
df.show()
```

```
df.printSchema()
```

```
df.columns
```

```
### HAndling CAtegorical Features
```

```
from pyspark.ml.feature import StringIndexer
```

```
index= StringIndexer(inputCols=['sex','smoker','day','time'] , outputCols=['sex_indexed','smoker_indexed','day_indexed','time_indexed'])
```

```
df_index=index.fit(df).transform(df)
```

```
df_index.show()
```

```
from pyspark.ml.feature import VectorAssembler
```

```
df_index.columns
```

```
featureassembler=VectorAssembler (inputCols=['tip','sex_indexed','smoker_indexed','day_indexed','time_indexed','size'],outputCol="Independent Feature")
```

```
output= featureassembler.transform(df_index)
```

```
output.show()
```

```
output.select('Independent Feature').show()
```

```
final= output.select('Independent Feature','total_bill')
```

```
final.show()
```

```
from pyspark.ml.regression import LinearRegression

from pyspark.ml.regression import LinearRegression
```

```
##Test Train Split
```

```
train_data,test_data=final.randomSplit([0.75,0.25])
regressor=LinearRegression(featuresCol='Independent Feature', labelCol='total_bill')
regressor=regressor.fit(train_data)
```

```
regressor.coefficients
```

```
regressor.intercepts
```

```
predict=regressor.evaluate(test_data)
```

```
predict.predictions.show()
```

```
predict.r2, predict.meanAbsoluteError,predict.meanSquaredError
```
